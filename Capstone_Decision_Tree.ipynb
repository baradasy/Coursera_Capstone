```python
import pandas as pd
import numpy as np
import itertools

import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
from matplotlib.ticker import NullFormatter
%matplotlib inline

from sklearn import preprocessing
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn import metrics
```

```python
mf = pd.read_csv('m_full.csv')
```

```python
mf.head()
```

```python
# remove Christmas series
mf = mf[~mf.series.isin(['yahx','btx','yahox'])]

#fix country spellings
mf.loc[mf.country=='Macao', 'country'] = 'Macau'
mf.loc[mf.country=='Trinidad & Tobago', 'country'] = 'Trinidad and Tobago'
```

```python
countries=mf.country.unique()
len(countries)
```

```python
mf[mf.location_type=='special'].shape
```

```python
m_tree = mf[['country','year','series']]
```

```python
m_tree[0:5]
```

```python
from sklearn import preprocessing

le_country = preprocessing.LabelEncoder()
le_country.fit(countries)
m_tree['country_code'] = le_country.transform(m_tree['country']) 

m_tree.head()
```

```python
X = m_tree[['country_code','year']].values
y = m_tree['series']
```

```python
X[0:5]
```

```python
y[0:5]
```

```python
X_train, X_test, y_train, y_test = train_test_split( X, y, test_size=0.3, random_state=3)
print ('Train set:', X_train.shape,  y_train.shape)
print ('Test set:', X_test.shape,  y_test.shape)
```

```python
mugTree = DecisionTreeClassifier(criterion="entropy", max_depth = 4)
mugTree # it shows the default parameters
```

```python
mugTree.fit(X_train,y_train)
predTree = mugTree.predict(X_test)
```

```python
print (predTree [0:5])
print (y_test [0:5].values)
```

```python
print("Decision Tree's Accuracy: ", metrics.accuracy_score(y_test, predTree))
```
